```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(12, 12))
for j in range(2):
    for i in range(3):
        print(i)
        dfi = pd.read_csv(f"./CSV/1011_p{i}.csv")
        ax[j].plot(dfi[f'{j}'], '.', label=f"p{i}", markersize=0.5)
    ax[j].legend()
```

```python
import numpy as np
fig, ax = plt.subplots(2, 2, sharex=True, figsize=(18, 12))
fps = 24
for j in range(2):
    for i in range(3):
        dfi = pd.read_csv(f"./CSV/1011_p{i}.csv")
        t = np.arange(dfi.shape[0]) / fps
        ax[j][0].plot(t, dfi[f'{j}'], '.', label=f"p{i}", markersize=0.5)
    ax[j][0].legend()
ax[0][0].set_title("1011 (High)")

for j in range(2):
    for i in range(3):
        dfi = pd.read_csv(f"./CSV/1012_p{i}.csv")
        t = np.arange(dfi.shape[0]) / fps
        ax[j][1].plot(t, dfi[f'{j}'], '.', label=f"p{i}", markersize=0.5)
    ax[j][1].legend()

for i in range(2):
    ax[0][i].set_ylim(600, 1450)
    ax[0][i].set_ylabel("X")
    ax[1][i].set_ylim(200, 800)
    ax[1][i].set_ylabel("Y")

ax[1][0].set_xlabel("Time (s)")
ax[1][1].set_xlabel("Time (s)")
fig.suptitle("Nose")
ax[0][1].set_title("1012 (Low)")
fig.savefig("./Nose_compare_rawdata.jpg", dpi=300)
```
